```python
import pandas as pd
import statsmodels.api as sm
import matplotlib
import matplotlib.pyplot as plt
```

```python

# Load the data into a Pandas dataframe
data = pd.read_csv('DataSet.csv')

# Extract the predictor variables and the target variable
X = data[['TotalArea', 'Lowest.point', 'Pop2023']]
y = data['Maximum.elevation']

# Add a constant to the predictor variables
X = sm.add_constant(X)

# Fit the multiple linear regression model using Ordinary Least Squares (OLS)
model = sm.OLS(y, X).fit()
```

```python
# Create a scatter plot of the data
plt.scatter(data['TotalArea'], y)

# Add the regression line to the plot
plt.plot(data['TotalArea'], model.predict(X), color='red')

# Add axis labels and a title
plt.xlabel('TotalArea')
plt.ylabel('Maximum Elevation')
plt.title('Regression Line for Maximum Elevation')

# Save the plot to a file
plt.savefig('Multi2.png')

# Show the plot
plt.show()
```
